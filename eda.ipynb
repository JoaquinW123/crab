```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from matplotlib.colors import ListedColormap

crab_data = pd.read_csv('CrabAgePrediction.csv')
crab_data.tail()
```

```python
df = pd.get_dummies(crab_data, columns = ['Sex'])
df.columns
df.head()
```

```python
columns = ['Length','Diameter','Height','Weight','Shucked Weight', 'Viscera Weight', 'Shell Weight','Age','Sex_F','Sex_I','Sex_M']

#correlation matrix
corr = df[columns].corr()

#heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Crab Features and Age')
plt.tick_params(axis='x', rotation=25)
plt.show()
```

```python
# Scatterplot: Gas Price vs. Crude Oil Price (highlight outliers)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Weight', y='Age', data= crab_data, hue='Sex', alpha=0.5)
plt.title('Weight vs. Age (Outliers Determined By Density)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
